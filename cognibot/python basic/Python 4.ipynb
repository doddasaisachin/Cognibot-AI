The primary difference of plt.scatter from plt.plot is that it can be used to create scatter plots where the properties of each individual point (size, face color, edge color, etc.) can be individually controlled or mapped to data.

For large datasets, the difference between these two can lead to vastly different performance, and for this reason, plt.plot should be preferred over plt.scatter for large datasets.

```python
np.random.seed(19680801)

N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = (30 * np.random.rand(N))**2  # 0 to 15 point radii

plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.show()
```

# Subplots and Bar plots

```python
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

plt.figure(figsize=(9, 3))

plt.subplot(1, 3, 1) # Total 1 rows, 3 columns and this is the first plot
plt.bar(names, values)
plt.subplot(1, 3, 2)
plt.scatter(names, values)
plt.subplot(1, 3, 3)
plt.plot(names, values)
# plt.title("Third subplot")
plt.suptitle('Categorical Plotting')
plt.show()
```

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure(figsize=(8, 8))
plt.subplot(2, 1, 1)
plt.plot(t1, f(t1), 'bo', t2, f(t2),'k' )

plt.subplot(2, 1, 2)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```

### Multiple figures and axis

```python
plt.figure(1)                # the first figure
plt.subplot(2,1,1)             # the first subplot in the first figure
plt.plot([1, 2, 3])
plt.subplot(2, 1, 2)             # the second subplot in the first figure
plt.plot([4, 5, 6])


plt.figure(2)                # a second figure
plt.plot([8, 9, 10])         # creates a subplot(111) by default
plt.title("Figure 2")

plt.figure(1)                # figure 1 current; subplot(212) still current
plt.subplot(2, 1, 1)             # make subplot(211) in figure1 current
plt.scatter([0, 1, 2], [4, 5, 6]) # subplot 211 title

plt.show()
# abc.savefig('fig_name.png')
```

### Adding text

```python
# evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.2)

# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.xlabel('X Label')
plt.ylabel('Y Label', fontsize=14, color='red')
plt.title('Title')
plt.text(2, 40, 'Write Here')

plt.show()
```

```python
my_plot = plt.figure()
ax = plt.subplot(1,1,1)

t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)
line, = plt.plot(t, s, lw=2)

plt.annotate('local max', xy=(2, 1), xytext=(3, 1.5),
             arrowprops=dict(facecolor='green'),
             )

plt.ylim(-2, 2)
plt.show()
```
