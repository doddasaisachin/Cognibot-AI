```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv('heart.csv')
```

```python
df.shape
```

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
scale.fit(df)
df1 = pd.DataFrame(scale.transform(df), columns=df.columns)
df1.index = df.index
df1.head()
```

```python
df1.describe()
```

```python
df.isnull().sum()
```

```python
# Why drop target - otherwise it will be PC1
target = df1.pop('target')
```

```python
df1.shape
```

```python
model = PCA(n_components=2)
model.fit(df1)
model.transform(df1)
```

```python
model.explained_variance_ratio_
```

```python
model.explained_variance_ratio_.sum()
```

```python
model.components_
```

```python
model2 = PCA()
model2.fit(df1)
model2.transform(df1)
```

```python
model2.explained_variance_ratio_
```
